```python
df.head(7)
```

3. Tampilkan informasi dataframe mengenai nama-nama kolom, jumlah data yang terisi, dan tipe datanya

4. Tampilkan karyawan bernama Rosemary Lee dan pekerjaannya

```python
df.loc[df.EmployeeName == 'Rosemary Lee', ['EmployeeName', 'JobTitle']]
```

5. Tampilkan nama karyawan dengan TotalPayBenefits paling besar

```python
df.loc[df.TotalPayBenefits == df.TotalPayBenefits.max(), 'EmployeeName'].values[0].capitalize()
```

6. Tampilkan dataset diurutkan dari tahun paling tua, kemudian dari BasePay paling besar

```python
df.sort_values(by = ['Year', 'BasePay'],
               ascending = [True, False]).reset_index(drop = True)
```

7. Tampilkan missing values pada kolom BasePay dan OvertimePay

```python
display(df[df.BasePay.isna()])
display(df[df.OvertimePay.isna()])
```

8. Tampilkan jumlah baris dari dataframe df

```python
print(f'Jumlah baris df : {len(df)}')
```

9. Tampilkan karyawan dengan BasePay lebih dari 300 ribu dollar dan OvertimePay-nya bukan 0 dollar

```python
df[(df.BasePay > 300000) & (df.OvertimePay != 0)]
```

10. Tampilkan data karyawan yang selisih TotalPayBenefits dan BasePay-nya bernilai negatif

```python
df[(df.TotalPayBenefits - df.BasePay < 0) | (df.BasePay - df.TotalPayBenefits < 0)]
```

11. JobTitle apa saja yang hanya muncul 1 kali (hanya dimiliki oleh 1 orang)?

```python
# df['JobTitle'].value_counts()[df['JobTitle'].value_counts() == 1]
temp_count = df['JobTitle'].value_counts()
temp_count[temp_count == 1]
```

12. Berapa median dari TotalPayBenefits pada tahun 2011

```python
print(f'Median dari TotalPayBenefits di tahun 2011 : {df[df.Year == 2011].TotalPayBenefits.median()}')
```
